```python
#Dependencies
import pandas as pd

#Load the Files
school_data_csv = "Resources/schools_complete.csv"
student_data_csv = "Resources/students_complete.csv"

#Read the files and store in DataFrames
school_data = pd.read_csv(school_data_csv)
student_data = pd.read_csv(student_data_csv)

#Merge the DataFrames on school name
school_data_merged = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_merged.head()
```

```python
#Count the Schools
school_count = len(school_data_merged["school_name"].unique())
school_count
```

```python
#Count the students
student_count= len(school_data_merged["student_name"].unique())
student_count
```

```python
#Total the Unique School Budgets
budget_total=school_data_merged["budget"].unique().sum()
budget_total
```

```python
#Find the average who passed math
avg_math=school_data_merged["math_score"].mean()
avg_math
```

```python
#Find the average who passed reading
avg_read=school_data_merged["reading_score"].mean()
avg_read
```

```python
#Count those who passed math, then find the percentage passing math
passing_math=school_data_merged.loc[school_data_merged["math_score"]>=70,:]
percentage_passing_math=(len(passing_math)/len(school_data_merged))*100
percentage_passing_math
```

```python
#Count those who passed reading, then find the percentage passing reading
passing_reading=school_data_merged.loc[school_data_merged["reading_score"]>=70,:]
percentage_passing_reading=(len(passing_reading)/len(school_data_merged))*100
percentage_passing_reading
```

```python
#Count those who passed both math and reading, then find percentage passing overall
overall_passing=school_data_merged.loc[(school_data_merged["reading_score"]>=70) &
(school_data_merged["math_score"]>=70),:]
overall_percentage_passing=(len(overall_passing)/len(school_data_merged))*100
overall_percentage_passing
```

```python
#Create data frame summarizing the district information

district_summary_df = pd.DataFrame({"Total Schools": [school_count],
                              "Total Students": student_count,
                              "Total Budget": budget_total,
                              "Average Math Score": avg_math,
                              "Average Reading Score": avg_read,
                              "Percentage Passing Math": percentage_passing_math,
                              "Percentage Passing Reading": percentage_passing_reading,
                              "Percentage Passing Overall": overall_percentage_passing})
district_summary_df
```

```python
#Set the index
school_types=school_data.set_index(["school_name"])["type"]
school_name_df=school_data_merged.groupby(["school_name"])
school_name_df.head()
```

```python
#Find student count for each school
school_student_count=school_name_df["Student ID"].count()
school_student_count
```

```python
#Find the budget for each school
school_budget=school_name_df["budget"].mean()
school_budget
```

```python
#Find the budget per student
student_budget=(school_budget/school_student_count)
student_budget
```

```python
#Average Math Score by school
school_avg_math=school_name_df["math_score"].mean()
school_avg_math
```

```python
#Average Reading Score by school
school_avg_reading=school_name_df["reading_score"].mean()
school_avg_reading
```

```python
#Percent of students passing math by school
school_pass_math=school_data_merged.loc[school_data_merged["math_score"]>=70]
school_pass_math_per=school_pass_math.groupby(["school_name"]).count()["student_name"]/school_student_count*100
school_pass_math_per


# passing_math=school_data_merged.loc[school_data_merged["math_score"]>=70,:]
# percentage_passing_math=(len(passing_math)/len(school_data_merged))*100
# percentage_passing_math
```

```python
#Percent of students passing reading by school
school_pass_read=school_data_merged.loc[school_data_merged["reading_score"]>=70]
school_pass_read_per=school_pass_read.groupby(["school_name"]).count()["student_name"]/school_student_count*100
school_pass_read_per
```

```python
#Percent Passing Overall by School
school_pass_overall=school_data_merged.loc[(school_data_merged["math_score"]>=70) & (school_data_merged["reading_score"]>=70)]
school_pass_overall_per=school_pass_overall.groupby(["school_name"]).count()["student_name"]/school_student_count*100
school_pass_overall_per
```

```python
school_summary_df = pd.DataFrame({"School Type": school_types,
                              "Total Students": school_student_count,
                              "Total School Budget": school_budget,
                              "Per Student Budget": student_budget,
                              "Average Math Score": school_avg_math,
                              "Average Reading Score": school_avg_reading,
                              "Percentage Passing Math": school_pass_math_per,
                              "Percentage Passing Reading": school_pass_read_per,
                              "Percentage Passing Overall": school_pass_overall_per
                              })
school_summary_df

```

```python
#Top Performing Schools by % Overall Passing
top_performing=school_summary_df.sort_values("Percentage Passing Overall",ascending=False)
top_performing.head()
```

```python
#Bottom Performing Schools by % Overall Passing
bottom_performing=school_summary_df.sort_values("Percentage Passing Overall")
bottom_performing.head()
```

```python
#Scores by School Spending
school_spending_df=school_summary_df
#Create the Bins
bins=[0,585,630,645,680]
bin_names=["<$585","585-630","$630-645","645-680"]

#Slice the data and place it into bins
pd.cut(school_spending_df["Per Student Budget"],bins,labels=bin_names).head()
```

```python
#Add new column to School Spending
school_spending_df["Per Student Budget"]=pd.cut(school_spending_df["Per Student Budget"],bins,labels=bin_names)
school_spending_df.head()
```

```python
spending_math=school_spending_df.groupby(["Per Student Budget"]).mean()["Average Math Score"]
spending_read=school_spending_df.groupby(["Per Student Budget"]).mean()["Average Reading Score"]
spending_math_per=school_spending_df.groupby(["Per Student Budget"]).mean()["Percentage Passing Math"]
spending_read_per=school_spending_df.groupby(["Per Student Budget"]).mean()["Percentage Passing Reading"]
spending_overall_per=school_spending_df.groupby(["Per Student Budget"]).mean()["Percentage Passing Overall"]
```

```python
scores_spending_df = pd.DataFrame({"Average Math Score": spending_math,
                              "Average Reading Score": spending_read,
                              "% Passing Math": spending_math_per,
                              "% Passing Reading": spending_read_per,
                              "Overall Passing Rate": spending_overall_per,
                              })
scores_spending_df
```
